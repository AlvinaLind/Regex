# Contact Verification and search - Using RegEx

## Email verification

```python
import re
import requests
```

```python
url = 'https://gist.githubusercontent.com/dinosaure/92552e7724f6057eedf6a1776a891c62/raw/17b17b1f0a4c5056a2531ec63673036faa60a92d/db.txt'
response = requests.get(url)
text = response.text
emails = text.splitlines()[:50]
emails[:20]
```

We know that the first part of the email adress can contain 
- Uppercase letters [A-Z]
- Lowercase letters [a-z]
- Numbers [0-9]

```python
pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,6}$' 
```

```python
for email in emails:
    status = "Valid" if re.search(pattern, email) else "NOT valid"
    print(f"{email:<40}  {status}")
```

```python
input_email = input('Enter an email address')
if re.search(pattern, input_email): 
    print('VALID email')
else: print('NOT valid email')
```

## Phone verification 

```python
import re

user_input = input("Skriv telefonnummer: ")
user_input = re.sub(r'[^0-9+]', '', user_input)

result = re.sub(r'^\s*\+46', '0', user_input)

result
```
